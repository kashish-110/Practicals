```python
from h2o.automl import H2OAutoML
```

```python
h2o.init()
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
data=h2o.import_file('/content/drive/MyDrive/train.csv')
```

```python
target_column = "Cabin"
feature_columns=data.columns
feature_columns.remove(target_column)
```

```python
train, test=data.split_frame(ratios=[0.8], seed=1234)
```

```python
train
```

```python
test
```

```python
aml=H2OAutoML(max_models=20, seed=1234, max_runtime_secs=3600, stopping_metric='AUC')
aml.train(x=feature_columns, y=target_column, training_frame=train)
```

```python
lb=aml.leaderboard
print(lb.head())
```

```python
best_model=aml.leader
predictions = best_model.predict(test)
print(predictions.head())
```

```python
performance=best_model.model_performance(test)
print(performance)
```

```python
from tpot import TPOTClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
```

```python
# Load dataset
data = load_iris()
X = data.data
y = data.target

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75, test_size=0.25, random_state=42)
```

```python
# Initialize TPOTClassifier
tpot = TPOTClassifier(generations=5, population_size=50, verbosity=2, random_state=42)

# Train the models
tpot.fit(X_train, y_train)

# Evaluate on the test data
accuracy=tpot.score(X_test, y_test)
```

```python
# Export the best pipeline to a Python file
tpot.export('best_model_pipeline.py')
```
